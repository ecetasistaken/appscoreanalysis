## Load and Merge data to print Detailed Informations

```python
import pandas as pd

# Load pre-COVID and post-COVID data
pre_covid_data = pd.read_csv('pre_covid_data(1).csv')
post_covid_data = pd.read_csv('covid_data(1).csv')

# Combine the data for a unified analysis
data = pd.concat([pre_covid_data, post_covid_data])

# Display the first few rows of the combined data
print(data.head())
```

 ## Data Cleaning and Preparation

```python
# Handling missing values
data = data.dropna(subset=['Size', 'Rating'])  # Dropping rows with missing Size or Rating

# Convert 'Reviews' to integer
data['Reviews'] = data['Reviews'].astype(int)

# Ensure 'Installs' is a string and then convert to integer (removing commas)
data['Installs'] = data['Installs'].astype(str).str.replace(',', '').astype(int)

# Ensure 'Price' is a string and then convert to float (removing $)
data['Price'] = data['Price'].astype(str).str.replace('$', '').astype(float)

# Convert 'Last Updated' to datetime
data['Last Updated'] = pd.to_datetime(data['Last Updated'])

# Convert 'Review Date' to datetime
data['Review Date'] = pd.to_datetime(data['Review Date'], format='%d-%m-%Y')

# Display cleaned data
print("\nCleaned dataset:")
print(data.head())

data.loc[:, 'Reviews'] = data['Reviews'].astype(int)
data.loc[:, 'Installs'] = data['Installs'].astype(str).str.replace(',', '').astype(int)
data.loc[:, 'Price'] = data['Price'].astype(str).str.replace('$', '').astype(float)
data.loc[:, 'Last Updated'] = pd.to_datetime(data['Last Updated'])
data.loc[:, 'Review Date'] = pd.to_datetime(data['Review Date'], format='%d-%m-%Y')
```

## Clustering Analysis

### K-Means Clustering

```python
from sklearn.cluster import KMeans

# Function to plot K-Means Clustering
def plot_kmeans(data, num_clusters):
    X = data[['Size', 'Rating']].dropna()
    kmeans = KMeans(n_clusters=num_clusters, random_state=0).fit(X)
    data['Cluster'] = kmeans.labels_
    
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=X['Size'] / 1e6, y=X['Rating'], hue=data['Cluster'], palette='viridis')
    plt.title(f'K-Means Clustering of App Size and Rating with {num_clusters} Clusters')
    plt.xlabel('Size (MB)')
    plt.ylabel('Rating')
    plt.show()

# Plot K-Means with different number of clusters
plot_kmeans(data, 3)
plot_kmeans(data, 4)
plot_kmeans(data, 5)
```

### Agglomerative Clustering

```python
from sklearn.cluster import AgglomerativeClustering

# Function to plot Agglomerative Clustering
def plot_agglomerative(data, num_clusters):
    X = data[['Size', 'Rating']].dropna()
    agglomerative = AgglomerativeClustering(n_clusters=num_clusters)
    labels = agglomerative.fit_predict(X)
    data['Cluster'] = labels
    
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=X['Size'] / 1e6, y=X['Rating'], hue=data['Cluster'], palette='viridis')
    plt.title(f'Agglomerative Clustering of App Size and Rating with {num_clusters} Clusters')
    plt.xlabel('Size (MB)')
    plt.ylabel('Rating')
    plt.show()

# Plot Agglomerative Clustering with different number of clusters
plot_agglomerative(data, 3)
plot_agglomerative(data, 4)
plot_agglomerative(data, 5)
```

### DBSCAN Clustering

```python
from sklearn.cluster import DBSCAN

# Function to plot DBSCAN Clustering
def plot_dbscan(data, eps, min_samples):
    X = data[['Size', 'Rating']].dropna()
    dbscan = DBSCAN(eps=eps, min_samples=min_samples)
    labels = dbscan.fit_predict(X)
    data['Cluster'] = labels
    
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=X['Size'] / 1e6, y=X['Rating'], hue=data['Cluster'], palette='viridis')
    plt.title(f'DBSCAN Clustering of App Size and Rating (eps={eps}, min_samples={min_samples})')
    plt.xlabel('Size (MB)')
    plt.ylabel('Rating')
    plt.show()

# Plot DBSCAN with different parameters
plot_dbscan(data, eps=1e7, min_samples=5)  # Adjust eps and min_samples based on your data scale
plot_dbscan(data, eps=1e7, min_samples=10)
plot_dbscan(data, eps=2e7, min_samples=5)
```

```python
# Including more features: 'Size', 'Rating', 'Reviews', 'Installs', 'Price'
features = ['Size', 'Rating', 'Reviews', 'Installs', 'Price']
X = data[features].dropna()

# Standardize the features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply K-Means with standardized features
kmeans = KMeans(n_clusters=3, random_state=0).fit(X_scaled)
data['Cluster'] = kmeans.labels_

# Plot the clusters (using first two principal components for visualization)
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=data['Cluster'], palette='viridis')
plt.title('K-Means Clustering with More Features')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

### Analyzing Cluster Quality

```python
from sklearn.metrics import silhouette_score

# Use 'Size' and 'Rating' for clustering
X = data[['Size', 'Rating']].dropna()

# Apply K-Means with different number of clusters and calculate silhouette score
for num_clusters in range(2, 20):
    kmeans = KMeans(n_clusters=num_clusters, random_state=0).fit(X)
    labels = kmeans.labels_
    score = silhouette_score(X, labels)
    print(f'Silhouette Score for {num_clusters} clusters: {score}')
```

```python
 # Step 1: Train-Test Split

from sklearn.model_selection import train_test_split

# Select relevant features for clustering
features = ['Size', 'Rating', 'Reviews', 'Installs', 'Price']
data_clean = data[features].dropna()

# Split the data into train and test sets
X_train, X_test = train_test_split(data_clean, test_size=0.3, random_state=42)

print(f"Train set size: {X_train.shape[0]}")
print(f"Test set size: {X_test.shape[0]}")
```

```python
 # Step 2: Standardize the Features

from sklearn.preprocessing import StandardScaler

# Standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Step 3: Clustering
from sklearn.cluster import KMeans

# Determine the optimal number of clusters using the elbow method
inertia = []
for n in range(1, 11):
    kmeans = KMeans(n_clusters=n, random_state=42)
    kmeans.fit(X_train_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.show()

# Based on the elbow plot, choose the number of clusters (e.g., 3)
optimal_clusters = 3
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
kmeans.fit(X_train_scaled)

# Add cluster labels to the train set
X_train['Cluster'] = kmeans.labels_

# Show the number of points in each cluster
print(X_train['Cluster'].value_counts())
```

```python
# Step 4: Visualize Clusters

# Use PCA for 2D visualization
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train_scaled)

plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_train_pca[:, 0], y=X_train_pca[:, 1], hue=X_train['Cluster'], palette='viridis')
plt.title('Train Set Clusters')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
# Step 5: Predict Clusters for Test Set

# Predict clusters for the test set
test_clusters = kmeans.predict(X_test_scaled)

# Add cluster labels to the test set
X_test['Cluster'] = test_clusters

# Show the number of points in each cluster
print(X_test['Cluster'].value_counts())

# Visualize test set clusters
X_test_pca = pca.transform(X_test_scaled)

plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_test_pca[:, 0], y=X_test_pca[:, 1], hue=X_test['Cluster'], palette='viridis')
plt.title('Test Set Clusters')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
# Step 6: Evaluate Clustering Performance

from sklearn.metrics import silhouette_score

# Calculate silhouette score for train and test sets
train_silhouette_score = silhouette_score(X_train_scaled, X_train['Cluster'])
test_silhouette_score = silhouette_score(X_test_scaled, X_test['Cluster'])

print(f"Silhouette Score for Train Set: {train_silhouette_score}")
print(f"Silhouette Score for Test Set: {test_silhouette_score}")
```

```python
from sklearn.cluster import DBSCAN

# Using DBSCAN clustering for additional analysis
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan.fit(X_train_scaled)

# Add cluster labels to the train set
X_train['Cluster'] = dbscan.labels_

# Show the number of points in each cluster
print(X_train['Cluster'].value_counts())

# Visualize DBSCAN clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_train_pca[:, 0], y=X_train_pca[:, 1], hue=X_train['Cluster'], palette='viridis')
plt.title('DBSCAN Train Set Clusters')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
